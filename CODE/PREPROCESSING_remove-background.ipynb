## 去除外框並儲存


````python
## Outlines
x1 = 60; x2 = 260
y1 = 60; y2 = 290

data_bg_rm = {} # Dataset background removed


for key in data.keys():
    data_bg_rm[key] = []
    for img in data[key]:
        if len(img.shape) == 2:
            data_bg_rm[key].append(img[x1:x2, y1:y2])
````
